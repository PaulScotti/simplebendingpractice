The current code reproduces [this example](https://www.mathalino.com/reviewer/mechanics-and-strength-of-materials/solution-to-problem-448-relationship-between-load-shear).

````python
# Initialization (leave this section as it is, unless you know what you are doing)
%matplotlib inline
import os
import sympy
from beambending.beam import Beam, DistributedLoad, PointLoad
x = sympy.symbols("x")  # uncomment if you use variable distributed loads


# Go on and play around with the values in this section. See what happens!
beam = Beam(9)
beam.fixed_support = 2    # x-coordinate of the fixed support
beam.rolling_support = 7  # x-coordinate of the rolling support
beam.add_loads((
                PointLoad(-20, 3),  # 20kN downwards, at x=3m
                DistributedLoad(-10, (3, 9)),  # 10 kN/m, downward, for 3m <= x <= 9m
                DistributedLoad(-20, (0, 2)),  # 20 kN/m downwards, for 0m <= x <= 2m
))

# Output generation (leave this section as it is, unless you know what you are doing)
fig = beam.plot()
fig.savefig("./beamtest.pdf")
````
